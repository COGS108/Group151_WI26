## Research Question

To what extent does social media addiction severity predict academic performance and relationship satisfaction, and do demographic factors such as age, gender, and relationship status moderate these effects?

## Background and Prior Work

Social media platforms have become deeply embedded in daily life, especially for students, serving as tools for communication, entertainment, and social connection. As usage has increased, so have concerns about problematic or addictive patterns of use and their potential impact on academic and relational outcomes. Prior research has shown that increased social media use is associated with reduced face‑to‑face interaction and higher distraction among young adults (Twenge, 2019; Kuss & Griffiths, 2017). Heavy social media use has also been linked to procrastination, attentional fragmentation, and reduced time available for studying and in‑person interactions. 

Excessive social media use may harm academic performance by displacing study time, fragmenting attention, and encouraging multitasking. Students who spend more hours per day on social media may experience lower grades, reduced focus, and higher stress, especially when platforms are used during study sessions or late at night. Prior work by Junco (2012) found that frequent social media use is negatively associated with academic outcomes, supporting this pattern. In the dataset we use, students report both their average daily social media usage hours and whether social media affects their academic performance, allowing us to directly examine these relationships. 

Social media addiction may also strain relationships. Constant exposure to curated posts can foster unrealistic expectations and social comparison, potentially lowering satisfaction with one’s own friendships or romantic relationships. Research by Appel et al. (2016) suggests that social comparison on social media can reduce relationship satisfaction and increase feelings of inadequacy. In addition, time spent online can replace face‑to‑face interactions, which may increase loneliness and conflict with friends, family, or partners. The survey we analyze includes measures of relationship satisfaction and conflict, enabling us to test whether higher social media addiction severity is associated with poorer relationship outcomes. 

Finally, demographic factors such as age, gender, and education level may shape how social media addiction relates to academic and relational outcomes. For example, younger students or undergraduates might be more vulnerable to distraction, while gender differences in platform use and communication styles could influence relationship dynamics. Because the dataset includes information on age, gender, education level, and country, we can explore whether these factors moderate the associations between social media addiction, academic performance, and relationship satisfaction. However, because both our dataset and much of the prior literature rely on self‑reported, cross‑sectional data, causal interpretations should be made cautiously.
