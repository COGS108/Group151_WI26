## Research Question

To what extent does social media addiction severity predict academic performance and relationship satisfaction, and do demographic factors such as age, gender, and relationship status moderate these effects?

## Background and Prior Work

In our current time, social media has become increasingly prevalent as the world becomes more digital. Social media platforms have been ingrained in our daily lives as a means of communication, connecting, and socializing with friends, family, and peers. While these platforms facilitate connection, there has been a growing concern that heavy social media use has been connected to addictive behaviours and negative impacts on the individual and their relationships. We now see a clear reduction in face-to-face interactions, which can interfere with building friendships and relationships, and can impact mental health and academic performance through constant distraction. 

Social Media addiction can harm relationships and academic performance in many ways. Any excessive use of social media easily takes away from quality time spent with friends, loved ones, and studying. Constantly seeing curated posts from others can also create unrealistic expectations and persuade people to believe that their relationships are unsatisfactory. Instant gratification from likes or funny videos can overload dopamine levels, causing the individual to constantly seek more of such content, which can cause a nasty feedback loop, driving the user to stay on the platform or log in more frequently. Understanding these patterns can help derive strategies to combat social media addiction and help people adopt healthier habits and protect their academic and relationship success. 

## Hypothesis


We hypothesize that greater social media addiction will be associated with greater academic and relationship strain.

Academic Impact: There is a statistically significant negative correlation between daily time spent on social media and academic performance. We predict that for every hour increase in daily usage, there will be a measurable decrease in grades because of procrastination and reduced study hours.
- Lower GPA
- Reduced study time 
- Increased procrastination
- Lower focus
- Higher stress

Relationship strain: we predict that individuals with higher social media addiction will report lower relationship satisfaction and a higher frequency of conflict with friends and family.
Lower relationship satisfaction with friends, family, and partners
- Increased frequency of conflicts
- Reduction in in-person interactions
- Higher levels of jealousy
- More lonely

Attention and Cognitive control : We hypothesize that there is a positive    correlation between excessive social media reliance and decreased attention span, we suggest that the intermittent reinforcement from social media leads to constant dopamine seeking behavior, creating a barrier to sustained concentration on academic performance or some kind of task.
- Reduced impulse control
- Decreased attention span
- Brain memory retention issues
