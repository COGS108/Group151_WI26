## Research Question

To what extent does social media addiction severity predict academic performance and relationship satisfaction, and do demographic factors such as age, gender, and relationship status moderate these effects?

## Background and Prior Work

Social media has become deeply embedded in the daily lives of students, serving as a primary tool for communication, entertainment, and social connection. As usage has increased, researchers have raised concerns about the potential for problematic or addictive patterns of engagement. Excessive social media use has been linked to reduced face‑to‑face interaction, increased distraction, and difficulties maintaining focus on academic tasks. 

Prior work suggests that heavy social media use may negatively affect academic performance by displacing study time, encouraging multitasking, and fragmenting attention. Students who spend more hours per day on social media often report lower grades, higher stress, and greater difficulty sustaining concentration. These patterns align with theories of attentional overload and reward‑driven behavior, where constant notifications and algorithmic reinforcement make it harder to disengage. 

Social media addiction has also been associated with relationship strain. Exposure to curated online content can heighten social comparison, reduce satisfaction with one’s own relationships, and increase feelings of jealousy or inadequacy. Additionally, time spent online may replace in‑person interactions, potentially weakening friendships, family relationships, or romantic partnerships. Studies have found that individuals with higher levels of social media addiction report lower relationship satisfaction and more frequent interpersonal conflict. 

Demographic factors may further shape these outcomes. Age, gender, and education level can influence both how students use social media and how vulnerable they are to its negative effects. For example, younger students may be more susceptible to distraction, while gender differences in communication styles may affect relationship dynamics. Because our dataset includes demographic variables alongside measures of social media usage, academic impact, and relationship satisfaction, it allows us to examine not only the direct associations but also how these relationships vary across groups.
