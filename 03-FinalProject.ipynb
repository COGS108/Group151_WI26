## Research Question

To what extent does social media addiction severity in college students predict academic performance and relationship satisfaction, and how do demographic factors such as age, gender, and relationship status moderate these effects?

## Background and Prior Work

As the world becomes more digital, social media has become increasingly prevalent in forms of communication, entertainment, and social connection. But as it becomes more prevalent, so has the concerns on people’s problematic/addictive usage of social media. These concerns include social media’s effect on an individual’s mental health, relationships, and academic performance. 

Prior research has shown that increased social media use is associated with higher distraction among young adults (Twenge, 2019; Kuss & Griffiths, 2017). Not to mention, excessive social media usage is also linked to attention fragmentation, procrastination, and reduced study time. All of which can harm academic performance with potential results such as  multitasking, lower grades and higher stress. Prior work by Junco (2012) found that frequent social media use is negatively associated with academic outcomes, supporting this pattern. In the dataset, students report both their average daily social media usage hours and whether social media affects their academic performance, allowing us to directly examine these relationships.

Social media addiction may also strain relationships. Constant exposure to curated posts can foster unrealistic expectations, which can lower satisfaction in your friendships or romantic relationships. Research by Appel et al. (2016) suggests that social comparison on social media can reduce relationship satisfaction and increase feelings of inadequacy. In addition, time spent online can replace face‑to‑face interactions, which may increase loneliness and conflict with friends, family, or partners. The survey we analyze includes measures of relationship satisfaction and conflict, enabling us to test whether higher social media addiction severity is associated with poorer relationship outcomes.

Finally, demographic factors such as age, gender, and education level may shape how social media addiction relates to academic and relational outcomes. For example, younger students or undergraduates might be more vulnerable to distraction, while gender differences in platform use and communication styles could influence relationship dynamics. Because the dataset includes information on age, gender, education level, and country, we can explore whether these factors moderate the associations between social media addiction, academic performance, and relationship satisfaction. However, because both our dataset and much of the prior literature rely on self‑reported, cross‑sectional data, causal interpretations should be made cautiously.

## Hypothesis


We hypothesize that higher levels of social media addiction will be associated with lower academic performance and lower relationship satisfaction among college students.

Academic Performance: Students with higher social media addiction severity and greater daily social media usage will report lower academic performance. Rationale: Prior research shows that heavy social media use reduces study time, increases multitasking, and disrupts sustained attention (Junco, 2012; Kuss & Griffiths, 2017). These mechanisms suggest that students who spend more time on social media may experience lower grades, reduced focus, and higher academic strain.

Relationship Satisfaction: Higher social media addiction severity will be associated with lower relationship satisfaction and more frequent interpersonal conflict. Rationale: Social comparison, jealousy, and unrealistic expectations created by curated online content have been shown to reduce relationship satisfaction (Appel et al., 2016). Additionally, time spent online may replace in‑person interactions, increasing loneliness and conflict with friends, family, or partners.

Moderation by Demographics: The strength of the associations between social media addiction and (a) academic performance and (b) relationship satisfaction will vary by demographic factors such as age, gender, and education level. Rationale: Younger students may be more vulnerable to distraction, and gender differences in communication styles and platform use may influence how social media affects academic and relational outcomes. Because our dataset includes age, gender, and education level, we can test whether these factors moderate the observed relationships.

**Parsing Numbers from String Outputs in Numerical Columns**

*Parsing Numbers Function*

```python
import re

def parse_numeric_first(val):
    # grabs the first number it finds in a string, returns nan if nothing found
    if pd.isna(val):
        return np.nan
    s = str(val).strip().replace(',', '.')
    m = re.search(r'\d+\.?\d*', s)
    return float(m.group()) if m else np.nan
```

We first make a function to extract number that extracts the number for any digits, any they take a dot if there is any and then take a extra decimals behind the dot if there are any. We changed the , to . for in case of any mistakes. Default value is np.nan if nothing can be found.

**Parsing the Numbers**

```python
# parse the messy text columns to numbers
df2['clean_smh'] = df2['raw_smh'].apply(parse_numeric_first)
df2['clean_hours'] = df2['raw_hours'].apply(parse_numeric_first)
df2['clean_attend_percent'] = df2['raw_attend'].apply(parse_numeric_first)

# gpa is also messy - try to pull out a number
df2['gpa_numeric'] = df2['gpa'].apply(parse_numeric_first)
```

We chose to parse the numbers by applying the functions onto the old column into the new row.
