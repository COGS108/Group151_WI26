## Research Question

To what extent does social media addiction severity in college students predict academic performance and relationship satisfaction, and how do demographic factors such as age, gender, and relationship status moderate these effects?


## Background and Prior Work

In our current time, social media has become increasingly prevalent as the world becomes more digital. Social media platforms have been ingrained in our daily lives as a means of communication, connecting, and socializing with friends, family, and peers. While these platforms facilitate connection, there has been a growing concern that heavy social media use has been connected to addictive behaviours and negative impacts on the individual and their relationships. We now see a clear reduction in face-to-face interactions, which can interfere with building friendships and relationships, and can impact mental health and academic performance through constant distraction.

Social Media addiction can harm relationships and academic performance in many ways. Any excessive use of social media easily takes away from quality time spent with friends, loved ones, and studying. Constantly seeing curated posts from others can also create unrealistic expectations and persuade people to believe that their relationships are unsatisfactory. Instant gratification from likes or funny videos can overload dopamine levels, causing the individual to constantly seek more of such content, which can cause a nasty feedback loop, driving the user to stay on the platform or log in more frequently. Understanding these patterns can help derive strategies to combat social media addiction and help people adopt healthier habits and protect their academic and relationship success.

Social media platforms have become deeply embedded in daily life, especially for students, serving as tools for communication, entertainment, and social connection. As usage has increased, so have concerns about problematic or addictive patterns of use and their potential impact on academic and relational outcomes. Prior research has shown that increased social media use is associated with reduced face‑to‑face interaction and higher distraction among young adults (Twenge, 2019; Kuss & Griffiths, 2017). Heavy social media use has also been linked to procrastination, attentional fragmentation, and reduced time available for studying and in‑person interactions. Excessive social media use may harm academic performance by displacing study time, fragmenting attention, and encouraging multitasking. Students who spend more hours per day on social media may experience lower grades, reduced focus, and higher stress, especially when platforms are used during study sessions or late at night. Prior work by Junco (2012) found that frequent social media use is negatively associated with academic outcomes, supporting this pattern. In the dataset we use, students report both their average daily social media usage hours and whether social media affects their academic performance, allowing us to directly examine these relationships.

Social media addiction may also strain relationships. Constant exposure to curated posts can foster unrealistic expectations and social comparison, potentially lowering satisfaction with one’s own friendships or romantic relationships. Research by Appel et al. (2016) suggests that social comparison on social media can reduce relationship satisfaction and increase feelings of inadequacy. In addition, time spent online can replace face‑to‑face interactions, which may increase loneliness and conflict with friends, family, or partners. The survey we analyze includes measures of relationship satisfaction and conflict, enabling us to test whether higher social media addiction severity is associated with poorer relationship outcomes.

Finally, demographic factors such as age, gender, and education level may shape how social media addiction relates to academic and relational outcomes. For example, younger students or undergraduates might be more vulnerable to distraction, while gender differences in platform use and communication styles could influence relationship dynamics. Because the dataset includes information on age, gender, education level, and country, we can explore whether these factors moderate the associations between social media addiction, academic performance, and relationship satisfaction. However, because both our dataset and much of the prior literature rely on self‑reported, cross‑sectional data, causal interpretations should be made cautiously.

## Hypothesis


We hypothesize that higher levels of social media addiction will be associated with lower academic performance and lower relationship satisfaction among college students.

Academic Performance: Students with higher social media addiction severity and greater daily social media usage will report lower academic performance. Rationale: Prior research shows that heavy social media use reduces study time, increases multitasking, and disrupts sustained attention (Junco, 2012; Kuss & Griffiths, 2017). These mechanisms suggest that students who spend more time on social media may experience lower grades, reduced focus, and higher academic strain.

Relationship Satisfactinon: Higher social media addiction severity will be associated with lower relationship satisfaction and more frequent interpersonal conflict. Rationale: Social comparison, jealousy, and unrealistic expectations created by curated online content have been shown to reduce relationship satisfaction (Appel et al., 2016). Additionally, time spent online may replace in‑person interactions, increasing loneliness and conflict with friends, family, or partners.

Moderation by Demographics: The strength of the associations between social media addiction and (a) academic performance and (b) relationship satisfaction will vary by demographic factors such as age, gender, and education level. Rationale: Younger students may be more vulnerable to distraction, and gender differences in communication styles and platform use may influence how social media affects academic and relational outcomes. Because our dataset includes age, gender, and education level, we can test whether these factors moderate the observed relationships.

```python
# correlation heatmap - lets see how all the numeric vars relate to eachother

numeric_vars = ['Age', 'Avg_Daily_Usage_Hours', 'Addicted_Score', 'Mental_Health_Score',
                'Sleep_Hours_Per_Night', 'Conflicts_Over_Social_Media', 'Affects_Academic_Performance']
corr_matrix = df[numeric_vars].corr().round(2)

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax, fmt='.2f')
ax.set_title('correlation matrix - dataset 1')
plt.show()

# printing the correlations with addiction
print('correlations with addicted score')
for var in numeric_vars:
    if var != 'Addicted_Score':
        r = corr_matrix.loc['Addicted_Score', var]
        print(f'  addicted_score & {var}: r={r:+.2f}')
```

- addiction and usage hours are also really correlated (r=+0.83) obviously if youre more addicted youre spending more time on it
- addiction and sleep have a decent negative correlation (r=-0.76) so more addicted students sleep less

overall these correlations are actually surprisingly strong for social science data. usually you dont see r values above 0.8 this often. the negative relationship between addiction and mental health is particularly strong

```python
# ok now bringing in dataset 2 which actually has GPA numbers
# gotta clean up the messy columns first since the raw data has stuff like "6hrs" instead of numbers

import re

def parse_numeric_first(val):
    # grabs the first number it finds in a string, returns nan if nothing found
    if pd.isna(val):
        return np.nan
    s = str(val).strip().replace(',', '.')
    m = re.search(r'\d+\.?\d*', s)
    return float(m.group()) if m else np.nan

# rename columns to something we can actually work with
df2 = df2.rename(columns={
    'Time spent in social media (hours)': 'raw_smh',
    'Study time (In Hours)': 'raw_hours',
    'Attendance rate (In Percentile)': 'raw_attend',
    'Last Academic Result (GPA/CGPA)': 'gpa',
    'Sleep Disturbance on Sleep Quality': 'sleep_disturbance',
    'Mood Modification Scale': 'mood_modification',
    'Anxiety Scale ': 'anxiety_score',
    'Depression Scale': 'depression_score',
    'Self-esteem Scale': 'self_esteem_score',
    'Social Media Distraction During Academic Activities ': 'academic_sm_distraction'
})

# parse the messy text columns to numbers
df2['clean_smh'] = df2['raw_smh'].apply(parse_numeric_first)
df2['clean_hours'] = df2['raw_hours'].apply(parse_numeric_first)
df2['clean_attend_percent'] = df2['raw_attend'].apply(parse_numeric_first)

# gpa is also messy - try to pull out a number
df2['gpa_numeric'] = df2['gpa'].apply(parse_numeric_first)

# theres some crazy outliers like gpa=2021 so we gotta filter those out
# most gpas should be on a 0-5 scale so just keep those
df2_gpa = df2.dropna(subset=['gpa_numeric', 'clean_smh', 'clean_hours'])
df2_gpa = df2_gpa[df2_gpa['gpa_numeric'] <= 5]
print(f'dataset 2 rows w valid gpa after removing outliers: {len(df2_gpa)} out of {len(df2)}')
print(f'gpa range: {df2_gpa["gpa_numeric"].min():.2f} to {df2_gpa["gpa_numeric"].max():.2f}')

fig, axes = plt.subplots(1, 2, figsize=(11, 4))

# scatter - social media hours vs gpa
axes[0].scatter(df2_gpa['clean_smh'], df2_gpa['gpa_numeric'], alpha=0.3, s=20)
z1 = np.polyfit(df2_gpa['clean_smh'], df2_gpa['gpa_numeric'], 1)
p1 = np.poly1d(z1)
xs1 = np.linspace(df2_gpa['clean_smh'].min(), df2_gpa['clean_smh'].max(), 50)
axes[0].plot(xs1, p1(xs1), 'r-')
axes[0].set_title('social media hrs vs gpa')
axes[0].set_xlabel('social media hours')
axes[0].set_ylabel('gpa')

#scatter - study hours vs gpa
axes[1].scatter(df2_gpa['clean_hours'], df2_gpa['gpa_numeric'], alpha=0.3, s=20, color='purple')
z2 = np.polyfit(df2_gpa['clean_hours'], df2_gpa['gpa_numeric'], 1)
p2 = np.poly1d(z2)
xs2 = np.linspace(df2_gpa['clean_hours'].min(), df2_gpa['clean_hours'].max(), 50)
axes[1].plot(xs2, p2(xs2), 'r-')
axes[1].set_title('study hrs vs gpa')
axes[1].set_xlabel('study hours')
axes[1].set_ylabel('gpa')

fig.suptitle('dataset 2 supplementary analysis')
plt.tight_layout()
plt.show()

# correlation heatmap for dataset 2
d2_numeric = ['clean_smh', 'clean_hours', 'clean_attend_percent', 'sleep_disturbance',
              'mood_modification', 'anxiety_score', 'depression_score',
              'self_esteem_score', 'academic_sm_distraction', 'gpa_numeric']

corr2 = df2_gpa[d2_numeric].corr().round(2)

fig2, ax2 = plt.subplots(figsize=(9, 7))
sns.heatmap(corr2, annot=True, cmap='coolwarm', ax=ax2, fmt='.2f',
            annot_kws={'size': 8})
ax2.set_title('correlation matrix - dataset 2')
plt.tight_layout()
plt.show()

# key correlations w gpa
print()
print('correlations with gpa (dataset 2)')
for var in d2_numeric:
    if var != 'gpa_numeric':
        r = corr2.loc['gpa_numeric', var]
        print(f'  gpa & {var}: r={r:+.2f}')
```

- **social media hours vs gpa** - basically no correlation at all (r=-0.01). the scatter plot is just a blob with a flat trendline. so in this dataset at least, time on social media doesnt really predict gpa directly
- **study hours vs gpa** - theres a very weak positive correlation (r=+0.12) so studying more is slightly associated with higher gpa but its not a strong relationship. the trendline has a slight upward slope but theres a lot of scatter
- the strongest predictor of gpa in this dataset is actually **attendance** (r=+0.30) which makes sense, showing up to class helps lol
